```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8n.yaml') # Usamos o .yaml para inicializar com 4 canais

# Inicia o treinamento
model.train(
    data='galaxy_data.yaml', 
    epochs=100, 
    imgsz=640, 
    device=0 # 0 para GPU, 'cpu' se não tiver
)
```
